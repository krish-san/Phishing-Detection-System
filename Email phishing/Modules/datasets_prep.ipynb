```python
import pandas as pd

file_path = "data/urls/phishtank-dataset.csv"

phish_df = pd.read_csv(file_path, usecols=['url'], encoding='latin1', on_bad_lines='skip')
phish_df = phish_df.dropna().drop_duplicates()
phish_df['label'] = 1

phish_df.to_csv("data/urls/phishtank_cleaned.csv", index=False)

print(f"Loaded {len(phish_df)} phishing URLs.")
phish_df.head()
```

```python
from tranco import Tranco

tranco = Tranco(cache=True)
tlist = tranco.list()
top1m = tlist.top(1000000)

legit_df = pd.DataFrame({
    'url': ['http://' + domain for domain in top1m],
    'label': 0
})

legit_df = legit_df.sample(n=len(phish_df), random_state=42).reset_index(drop=True)
legit_df.to_csv("data/urls/tranco_legit.csv", index=False)
print("Legitimate URLs saved")
legit_df.head()
```
